```python
import pandas as pd

ride_data = "raw_data/ride_data.csv"
city_data = "raw_data/city_data.csv"

ride_df = pd.read_csv(ride_data)
city_df = pd.read_csv(city_data)


ride_df.columns
```

```python
# summary df by city

ride_summary_df = pd.DataFrame({
    "Total Number of Rides (Per City)" : ride_df.groupby(["city"]).count()["ride_id"],
    "Average Fare ($)": ride_df.groupby(["city"]).mean()["fare"].map("${:,.2f}".format)
    
})

ride_summary_df.head()
```

```python
city_df["Total Number of Rides (Per City)"] = ride_df.groupby(["city"]).count()["ride_id"]
city_df
```

```python
# passing in a df to plt

import matplotlib.pyplot as plt
import pandas as pd

x = [1,2,3,4,5]
y = [100, 321, 141, 132, 314]
z = [50,50,1000,50,50]

xyz_df = pd.DataFrame({
    "x": x,
    "y": y,
    "z": z
})

xyz_df

plt.scatter(xyz_df["x"],xyz_df["y"], s=xyz_df["z"])
plt.show()
```
